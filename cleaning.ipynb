```python
import pandas as pd 
```

```python
df = pd.read_csv('./data_scraping_10feb_raw.csv')
```

```python
df.head()    
```

```python
most_common = df['Actual Place Name'].value_counts()
print(most_common)
```

```python
# Source - https://stackoverflow.com/q/31269216
# Posted by rhino, modified by community. See post 'Timeline' for change history
# Retrieved 2026-02-12, License - CC BY-SA 3.0

df['Actual Place Name'] = list(map(str.upper, df['Actual Place Name']))
```

```python
df = df[~df['Actual Place Name'].isin(['TIDAK DITEMUKAN - NO RESULTS', 'ERROR: MULTIPLE RESULTS - GAGAL SEMUA KARTU'])]
```

```python
df
```

```python
most_common = df['Validasi'].value_counts()
print(most_common)
```

```python
df = df[~df['Validasi'].isin(['Tidak Ditemukan'])]
```

```python
df
```

```python

only_1 = df[df['Validasi'] == 'Ditemukan']
print(only_1)
```
